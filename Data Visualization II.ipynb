```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset from Seaborn
titanic_df = sns.load_dataset('titanic')

# Plot box plot for distribution of age with respect to gender and survival status
plt.figure(figsize=(10, 6))
sns.boxplot(data=titanic_df, x='sex', y='age', hue='survived')
plt.title('Distribution of Age with Respect to Gender and Survival Status')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.grid(True)
plt.savefig("age_distribution_gender_survival_boxplot.png")  # Save the plot as an image
plt.show()
```
