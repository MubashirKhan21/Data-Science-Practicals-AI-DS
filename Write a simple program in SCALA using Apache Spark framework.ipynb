```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import split, explode, count

# Initialize SparkSession
spark = SparkSession.builder \
    .appName("WordCount") \
    .master("local[*]") \
    .getOrCreate()

# Create DataFrame from a list of words
words_df = spark.createDataFrame([("hello world hello apache spark hello",)], ["text"])

# Split text into words
words_df = words_df.withColumn("words", split("text", " "))

# Explode array of words into separate rows
words_df = words_df.select(explode("words").alias("word"))

# Perform word count
word_counts_df = words_df.groupBy("word").agg(count("*").alias("count"))

# Show word counts
word_counts_df.show()

# Stop SparkSession
spark.stop()
```
